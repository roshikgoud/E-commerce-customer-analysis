```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
import seaborn as sns
import sklearn as sklearn

```

```python
df = pd.read_csv(r"C:\Users\roshi\OneDrive\Desktop\ecommerce_customer_behavior_dataset_v2.csv")
```

```python
df.head()
df.info()
df.describe()
print(df.isnull().sum())
print(df.duplicated().sum())
print(df['Order_ID'].duplicated().sum())
```

```python
df['Date']=pd.to_datetime(df['Date'])
df.info("Date")
```

```python
## revenue per order 
df['Revenue']=df['Total_Amount']
## avg price for each item
df['Avg_Price']=df['Total_Amount']/df['Quantity']
## extracting the time features 
df['Month']=df['Date'].dt.month
df['Day']=df['Date'].dt.day
```

Advance EDA 


```python
top_cat=df.groupby('Product_Category')['Revenue'].sum().sort_values(ascending=False)
print(top_cat)
```

```python
top_city=df.groupby('City')['Revenue'].sum().sort_values(ascending=False)
print(top_city)
```

```python
print(df['Payment_Method'].value_counts())
```

```python
print(df['Is_Returning_Customer'].value_counts())
```

```python
print(df.groupby('Delivery_Time_Days')['Customer_Rating'].mean())
```

Visualization


```python
plt.figure()
df.groupby('Product_Category')['Revenue'].sum().plot(kind='bar')
plt.title('Revenue by Product Category')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure()
df['Is_Returning_Customer'].value_counts().plot(kind='pie',autopct='%1.1f%%')
plt.title('Returning vs New Customers')
plt.show()
```

```python
plt.figure()
sns.lineplot(x='Delivery_Time_Days',y='Customer_Rating',data=df)
plt.title('Customer Rating vs Delivery Time')
plt.show()
```
